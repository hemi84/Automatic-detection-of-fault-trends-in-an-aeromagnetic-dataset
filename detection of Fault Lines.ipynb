```python
import cv2
import numpy as np
import pandas as pd
```

```python
import hemi_fault_detection as hf
```

```python
img = cv2.imread('ln_og.png')
```

```python
lines = hf.detect_fault(img)
```

```python
for points in lines:
    x1,y1,x2,y2 = points[0]
    cv2.line(img, (x1,y1),(x2,y2),(0,0,0))
```

```python
cv2.imwrite('faults.png', img)
```
